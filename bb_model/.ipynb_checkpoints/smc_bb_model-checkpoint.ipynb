# Beta-Binomial Model

We want to compute the posterior distribution for a proportion "p" in a yes/no setting (i. e., in a poll with n people interviewed, where they are asked whether they voted for candidate A or B, p could be the fraction of people voting for candidate A). 

The likelihood is the Binomial distribution:

$P_{n, y} = \binom{n}{y} p^y (1-p)^{n-y}$

In order to have an analytic form for the posterior (so as to be able to compare the ABC results with the ground truth), we choose a Beta prior on p:
    
$p \sim Beta(a, b)$

where the hyperparameters $a, b$ are specified below.

In this setting, we know that the posterior will also be a Beta distribution, with updated parameters:

$\pi(p|y, n, a, b) = Beta(a+y, b+n-y)$

## Generate data, parameters and weights at time t = 0

```python
size_data = 50
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 4.5))
ax.tick_params(labelsize=15)
np.random.seed(5)

p_true = 0.3
n_true = 20

y_true = np.random.binomial(n_true, p_true, size=size_data)

ax.hist(y_true, bins='auto', color='red')
ax.set_xlabel('number of successes', fontsize=15)
ax.set_ylabel('frequency', fontsize=15)

ax.set_title(f'$\mu$ = {np.mean(y_true):.3f} $-$ $\sigma$ = {np.std(y_true):.3f}', fontsize=15)
plt.tight_layout()
```

```python
y = np.sum(y_true)
y
```

```python
n_ = int(1e4)
p_th = 1.5e-4
interv = np.linspace(p_th, 1 - p_th, n_)

a, b = 0.5, 0.5  #hyperparameters
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 4))

ax.plot(interv, beta.pdf(interv, a, b), '-r', label='prior')
ax.plot(interv, beta.pdf(interv, a+y, b+size_data*n_true-y), '-b', label='analytic\nposterior')

ax.tick_params(labelsize = 18)

ax.set_xlabel('p', fontsize=18)
ax.set_ylabel('probability', fontsize=18)
plt.tight_layout()
plt.legend(loc=(0.4, 0.5), fontsize=18)
plt.show()
```

```python
mean_beta, var_beta = beta.stats(a+y, b+size_data*n_true-y, moments='mv')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(13, 5.5))
t = np.linspace(p_th, 1 - p_th, 2000)

alpha = 0.5

for n_round in [1, 4, 7, 20]:
    w_theta = pd.read_csv(path / f"w_theta_time_{n_round}_part_{n_part}.txt", sep=' ', names=['w', 'p'])
    ax[0].hist(w_theta['p'], density=True, bins='auto' , alpha=alpha, label=f'{n_round} SMC rounds')
    
ax[0].plot(t, beta.pdf(t, a+y, b+size_data*n_true-y), '--', color='black', alpha=1, linewidth=2.5, label='analytic\nposterior')
ax[0].set_title('ABC: $\mu$ = {:.3f}, $\sigma^2$ = {:.3e}\nGT: $\mu$ = {:.3f}, $\sigma^2$ = {:.3e}'
             .format(w_theta['p'].mean(), w_theta['p'].var(ddof=1), mean_beta, var_beta), fontsize=15)

ax[0].plot(interv, beta.pdf(interv, a, b), '-r', label='prior')
ax[0].plot(p_true*np.ones(50), np.linspace(0, max(beta.pdf(interv, a, b))), 'b-', label='true value')    
ax[1].hist(y_true, bins='auto', color='green', label='data')

y_labels = ['probability', 'frequency']
x_labels = ['p', '# successes']
loc = [(0.47, 0.25), 'best']
for i, ax_ in enumerate(ax):
    ax_.set_xlabel(x_labels[i], fontsize=20)
    ax_.set_ylabel(y_labels[i], fontsize=20)
    ax_.legend(loc=loc[i], fontsize=15)
    ax_.tick_params(labelsize = 18)
    
plt.tight_layout()
```
